```python
import geopandas as gpd
import osmium as osm
from shapely.geometry import MultiPolygon, Polygon


class OSMHandler(osm.SimpleHandler):
    def __init__(self):
        osm.SimpleHandler.__init__(self)
        self.polygons = []

    def area(self, a):
        if "admin_level" in a.tags and "name" in a.tags:
            rings = []
            for ring in a.outer_rings():
                rings.append(Polygon([(node.lon, node.lat) for node in ring]))

            if len(rings) > 1:
                geom = MultiPolygon(rings)
            else:
                geom = rings[0]

            self.polygons.append(
                {
                    "name": a.tags["name"],
                    "admin_level": a.tags["admin_level"],
                    "geometry": geom,
                }
            )


handler = OSMHandler()
handler.apply_file(r"C:\Users\Lenovo\Downloads\iowa-latest.osm.pbf", locations=True)

gdf = gpd.GeoDataFrame(handler.polygons, crs="EPSG:4326")

iowa_boundary = gdf[gdf["admin_level"] == "4"]

admin_centers = gdf[gdf["admin_level"] == "8"]

iowa_boundary.to_file(
    r"C:\Users\Lenovo\Downloads\iowa_boundary.geojson", driver="GeoJSON"
)
admin_centers.to_file(
    r"C:\Users\Lenovo\Downloads\admin_centers.geojson", driver="GeoJSON"
)

admin_center_names = admin_centers["name"].tolist()
print(admin_center_names)
```

```python
import geopandas as gpd
import osmium as osm
from shapely.geometry import MultiPolygon, Polygon


class OSMHandler(osm.SimpleHandler):
    def __init__(self):
        osm.SimpleHandler.__init__(self)
        self.polygons = []

    def area(self, a):
        if "admin_level" in a.tags and "name" in a.tags:
            rings = []
            for ring in a.outer_rings():
                rings.append(Polygon([(node.lon, node.lat) for node in ring]))

            if len(rings) > 1:
                geom = MultiPolygon(rings)
            else:
                geom = rings[0]

            self.polygons.append(
                {
                    "name": a.tags["name"],
                    "admin_level": a.tags["admin_level"],
                    "geometry": geom,
                }
            )


handler = OSMHandler()
handler.apply_file(r"C:\Users\Lenovo\Downloads\iowa-latest.osm.pbf")

gdf = gpd.GeoDataFrame(handler.polygons, crs="EPSG:4326")

gdf["geometry"] = gdf["geometry"].apply(
    lambda geom: MultiPolygon([geom]) if isinstance(geom, Polygon) else geom
)

iowa_boundary = gdf[(gdf["admin_level"] == "4") & (gdf["name"] == "Iowa")]

iowa_boundary_state = gdf[
    (gdf["admin_level"] == "6") & (gdf["name"].str.contains("County"))
]

iowa_boundary_region = gdf[(gdf["admin_level"] == "7")]

admin_centers = gdf[(gdf["admin_level"] == "8")]

iowa_boundary_state_gpkg = "iowa_boundary_state.gpkg"
iowa_boundary_region_gpkg = "iowa_boundary_region.gpkg"
iowa_boundary_gpkg = "iowa_boundary.gpkg"
admin_centers_gpkg = "admin_centers.gpkg"

iowa_boundary_state.to_file(iowa_boundary_state_gpkg, driver="GPKG")
iowa_boundary_region.to_file(iowa_boundary_region_gpkg, driver="GPKG")
iowa_boundary.to_file(iowa_boundary_gpkg, driver="GPKG")
admin_centers.to_file(admin_centers_gpkg, driver="GPKG")

iowa_boundary_state_csv = "iowa_boundary_state.csv"
iowa_boundary_region_csv = "iowa_boundary_region.csv"
iowa_boundary_csv = "iowa_boundary_only.csv"
admin_centers_csv = "admin_centers_only.csv"

iowa_boundary_state.to_csv(iowa_boundary_state_csv, index=False)
iowa_boundary_region.to_csv(iowa_boundary_region_csv, index=False)
iowa_boundary.to_csv(iowa_boundary_csv, index=False)
admin_centers.to_csv(admin_centers_csv, index=False)

admin_center_names = admin_centers["name"].tolist()
print(admin_center_names)
```
